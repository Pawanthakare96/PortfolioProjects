## From follwing code  we just comparing KNN imputation and Mean Imputation

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer

# Load the dataset
#credit_card = pd.read_csv("credit_missing.csv")

# Impute the missing values in the Annual_income column using mean
credit_card_mean = credit_card.copy()
mean_imputer = SimpleImputer(strategy='mean')

# Get the underlying NumPy array of the Annual_income column
annual_income_array = credit_card_mean['Annual_income'].values

# Reshape the Annual_income column to be 2D
annual_income_array = annual_income_array.reshape(-1, 1)

# Impute the missing values in the Annual_income column using mean
credit_card_mean['Annual_income'] = mean_imputer.fit_transform(annual_income_array)

# Impute the missing values in the Annual_income column using KNN
credit_card_knn = credit_card.copy()
knn_imputer = KNNImputer(n_neighbors=int(5))

# Get the underlying NumPy array of the Annual_income column
annual_income_array = credit_card_knn['Annual_income'].values

# Reshape the Annual_income column to be 2D
annual_income_array = annual_income_array.reshape(-1, 1)

# Impute the missing values in the Annual_income column using KNN
credit_card_knn['Annual_income'] = knn_imputer.fit_transform(annual_income_array)

```

## Treating outliers using IQR Method

```python
Q1=new_df['Annual_income'].quantile(0.25)
Q3=new_df['Annual_income'].quantile(0.75)
IQR=Q3-Q1

L_limit=Q1 - (1.5*IQR)
U_limit=Q3 + (1.5*IQR)
```

```python
new_df[(new_df['Annual_income']>U_limit) | (new_df['Annual_income']<L_limit)]
```

```python
new_df['Annual_income']=new_df['Annual_income'].clip(lower=L_limit,upper=U_limit)
```

# Train and Test Splititng of dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=9)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

# Logistic Regression Model

```python
# Modeling Building
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression(random_state=0)
model_LR.fit(x_train,y_train)

# Prediciton

prediction_train=model_LR.predict(x_train)
prediction_test=model_LR.predict(x_test)


```

```python
# Evaluation

from sklearn.metrics import accuracy_score

print('accuracy_score_train:',accuracy_score(y_train,prediction_train))
print('accuracy_score_test:',accuracy_score(y_test,prediction_test))

from sklearn.model_selection import cross_val_score

LR_cross_val_score=cross_val_score(model_LR,X,Y,cv=5,scoring='accuracy').mean()
LR_cross_val_score
```

```python
import sklearn.metrics as metrics

f1_score = metrics.f1_score(y_test, prediction_test, average='weighted')

print('f1_score:', f1_score)
```

# Applying KNN Algorithm

```python
# Modeling

from sklearn.neighbors import KNeighborsClassifier

model_KNN=KNeighborsClassifier()
model_KNN.fit(x_train,y_train)

## Prediction

prediction_test=model_KNN.predict(x_test)
prediction_train=model_KNN.predict(x_train)

## Evaluation

from sklearn.metrics import accuracy_score
print('accuracy_score:',accuracy_score(y_test,prediction_test))
print('accuracy_score:',accuracy_score(y_train,prediction_train))

from sklearn.model_selection import cross_val_score
print("cross_val-score:",cross_val_score(model_KNN,X,Y,cv=5,scoring="accuracy").mean())
```
